```python
# aT.b = a.b


a = np.random.randn(5)
b = np.random.randn(5)
c = np.random.randn(1) * a # makes a and b linearly dependent so angle between them will be 90 and cos(theta)=1

print(a, b, c, sep='\n')

aTb = np.dot(a, b)
aTc = np.dot(a, c)

norm_a = np.linalg.norm(a)
norm_b = np.linalg.norm(b)
norm_c = np.linalg.norm(c)

print(f'inequality: left side: {abs(aTb):.3f}, right side: {norm_a * norm_b:0.3f}')
print(f'equality:   left side: {abs(aTc):.3f}, right side: {norm_a * norm_c:0.3f}')
```
