# dot products

```python
v1 = np.array([1, 2, 3, 4, 5])
v2 = np.array([0, 4, -1, 5, 7])
```

```python
# performs a dot product for 1D arrays (vector)
'''
- If both `a` and `b` are 1-D arrays, it is inner product of vectors
  (without complex conjugation).

- If both `a` and `b` are 2-D arrays, it is matrix multiplication,
  but using :func:`matmul` or ``a @ b`` is preferred.
'''
dp1 = np.dot(v1, v2)
dp1
```

```python
# matrix multiplication for 2D arrays, and uses complex broadcasting for higher dimensions
# perform dot product operations between two matrices or a matrix and a vector.
dp2 = np.matmul(v1, v2)
dp3 = v1 @ v2
dp2, dp3
```

## dot products (associative, distributive)

```python
# a.(b+c) = a.b + a.c (is distributive)

n = 5 
a = np.random.randn(5)
b = np.random.randn(5)
c = np.random.randn(5)

res1 = np.dot(a, b+c)
res2 = np.dot(a, b) + np.dot(a, c)

res1, res2
```

```python
# a.(b.c) != (a.b).c (is not associative)

n = 5 
a = np.random.randn(5)
b = np.random.randn(5)
c = np.random.randn(5)

res1 = np.dot(a, np.dot(b, c))
res2 = np.dot(np.dot(a, b), c)

res1, res2
```

## dot products (commutative)

```python
# a.b = b.a (is commutative)

a = np.random.randn(100)
b = np.random.randn(100)

res1 = np.dot(a, b)
res2 = np.dot(b, a)

res1, res2, res1 - res2
```
