```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = np.array([3, -1])
B = np.array([2, 4])
```

```python
C = A + B
C
```

```python
plt.plot([0, A[0]], [0, A[1]], 'blue', label='A')
plt.plot([0, B[0]] + A[0], [0, B[1]] + A[1], 'red', label='B')
plt.plot([0, C[0]], [0, C[1]], 'green', label='C')

plt.axis('square')
plt.axis((-6, 6, -6, 6))
plt.plot([-6, 6], [0, 0], 'k--')
plt.plot([0, 0], [-6, 7], 'k--')
plt.grid()
plt.xticks(np.arange(-6, 7))
plt.yticks(np.arange(-6, 7))
plt.legend()
plt.show()
```
