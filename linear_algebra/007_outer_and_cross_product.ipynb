```python
import numpy as np
from PIL import Image
```

```python
a = np.array([1, 2, 3])
b = np.array([3, 2, 1])

print(f'dot product: \n{np.dot(a, b)}')
print(f'outer product:  \n{np.outer(a, b)}')
print(f'cross product: \n{np.cross(a, b)}')
```
