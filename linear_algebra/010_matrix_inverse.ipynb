#### (A × A⁻¹ = I)

```python
m = 4

A = np.random.randn(m, m)
A_inv = np.linalg.inv(A)

# A.A_inv = I (A × A⁻¹ = I)
I = A@A_inv
I
```

```python
np.isclose(I, 0)
```

```python
# simple plot

plt.subplot(131)
plt.imshow(A)
plt.title('Matrix A')

plt.subplot(132)
plt.imshow(A_inv)
plt.title('Matrix $A^{-1}$')

plt.subplot(133)
plt.imshow(I)
plt.title('$AA^{-1}$')
```

#### MCA Algorithm (Adjugate Formula)

M = Matrix of Minors (determinants of submatrices)

H = Sign matrix for Cofactors (checkerboard pattern of + and -)

C = Cofactor matrix (M × H element-wise)

A_inv = C.T / det(A) (transpose of cofactor matrix divided by determinant)

A⁻¹ = (1/det(A)) · adj(A)
where adj(A) = Cᵀ (transpose of cofactor matrix)

```python
m = 3
A = np.random.randn(m, m)

# minor matrix
M = np.zeros((m, m)) # minors matrix
H = np.zeros((m, m)) # sign matrix

for i in range(m):
    for j in range(m):
        rows = [True]*m
        rows[i] = False

        cols = [True]*m
        cols[j] = False

        M[i, j] = np.linalg.det(A[rows, :][:, cols])
        H[i, j] = (-1) ** (i+j)

C = H*M # cofactors matrix
A_inv = C.T/np.linalg.det(A)

result = A@A_inv

result, np.isclose(result, 0)
```

```python
# cleaner code 
"""
Adjugate Formula
A⁻¹ = adj(A) / det(A)
where adj(A) = Cᵀ (transpose of cofactor matrix)
"""
A = np.random.randn(3, 3)

m = A.shape[0]

# Matrix of minors
minors = np.zeros((m, m))

for i in range(m):
    for j in range(m):
        # Create submatrix by removing row i and column j
        rows = [r for r in range(m) if r != i]
        cols = [c for c in range(m) if c != j]
        minor = A[np.ix_(rows, cols)]
        minors[i, j] = np.linalg.det(minor)

# Sign matrix (+/-)
sign_matrix = np.array([[(-1) ** (i+j) for j in range(m)] for i in range(m)])

# Cofactor matrix = sign_matrix * minors
cofactors = sign_matrix * minors

# Adjugate = transpose of cofactor matrix
adjugate = cofactors.T

# Inverse = adjugate / determinant
det = np.linalg.det(A)

print('Matrix A:')
print(A)
print('\nA⁻¹ (Adjugate Formula):')
print(f'result: {adjugate / det}')
print('\nA⁻¹ (NumPy):')
print(np.linalg.inv(A))
```
