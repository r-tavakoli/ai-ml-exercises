```python
import numpy as np
import matplotlib.pyplot as plt
```

## Step 1: simple model

```python
# design matrix
data = np.array([-4, 0, -3, 1, 2, 8, 5, 8]).reshape(-1, 1)
n = len(data)
X = np.ones((n, 1))
X
```

```python
# fit the model
b = np.linalg.solve(X.T@X, X.T@data)
mu = np.mean(data)
b, mu
```

```python
# predict
y_hat = X@b
y_hat
```

```python
plt.plot(np.arange(1, n+1), data, color='blue', linestyle='-', marker='s', label='actual', alpha=0.7)
plt.plot(np.arange(1, n+1), y_hat, color='red', linestyle='--', marker='o', label='predict', alpha=0.7)
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

## Step 2: minimizing the distance

```python
# design matrix
data = np.array([-4, 0, -3, 1, 2, 8, 5, 8]).reshape(-1, 1)
n = len(data)
X = np.array(np.arange(0, n)).reshape(-1, 1)

# fit the model
b = np.linalg.solve(X.T@X, X.T@data)
mu = np.mean(data)
b, mu

# predict
y_hat = X@b
y_hat

X
```

```python
plt.plot(np.arange(1, n+1), data, color='blue', linestyle='-', marker='s', label='actual', alpha=0.7)
plt.plot(np.arange(1, n+1), y_hat, color='red', linestyle='--', marker='o', label='predict', alpha=0.7)
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

## Step 3: adding intercept

```python
# design matrix
data = np.array([-4, 0, -3, 1, 2, 8, 5, 8]).reshape(-1, 1)
n = len(data)
X = np.concatenate([np.ones((n, 1)), np.array(np.arange(0, n)).reshape(-1, 1)], axis=1)

# fit the model
b = np.linalg.solve(X.T@X, X.T@data)
mu = np.mean(data)
b, mu

# predict
y_hat = X@b
y_hat

X
```

```python
plt.plot(np.arange(1, n+1), data, color='blue', linestyle='-', marker='s', label='actual', alpha=0.7)
plt.plot(np.arange(1, n+1), y_hat, color='red', linestyle='--', marker='o', label='predict', alpha=0.7)
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```
