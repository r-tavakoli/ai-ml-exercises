```python
# 2D
# tetha = 90 
v1 = np.array([0, 1])
v2 = np.array([1, 0])
v3 = np.array([1, 0.5])
v4 = np.array([1, -1])
v5 = np.array([0, 0.5])

angle_radian_90 = np.arccos(np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2)))
angle_degree_90 = np.rad2deg(angle_radian_90)

angle_radian_less_than_90 = np.arccos(np.dot(v1, v3) / (np.linalg.norm(v1) * np.linalg.norm(v3)))
angle_degree_less_than_90 = np.rad2deg(angle_radian_less_than_90)

angle_radian_greater_than_90 = np.arccos(np.dot(v1, v4) / (np.linalg.norm(v1) * np.linalg.norm(v4)))
angle_degree_greater_than_90 = np.rad2deg(angle_radian_greater_than_90)

angle_radian_0 = np.arccos(np.dot(v1, v5) / (np.linalg.norm(v1) * np.linalg.norm(v5)))
angle_degree_0 = np.rad2deg(angle_radian_0)

#plot
fig = plt.figure()
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax4 = fig.add_subplot(224)

axlim = max([abs(max(v1)), abs(max(v2))])
axtick = round(axlim + 2)
# plt.xticks(np.arange(-axtick, axtick))
# plt.yticks(np.arange(-axtick, axtick))

# theta = 90
ax1.plot([0, v1[0]], [0, v1[1]], 'b', label='v1')
ax1.plot([0, v2[0]], [0, v2[1]], 'r', label='v2')
ax1.set_xticks(np.arange(-axtick, axtick))
ax1.set_yticks(np.arange(-axtick, axtick))
ax1.axis((-axlim, axlim, -axlim, axlim))
ax1.set_title(f'theta={angle_degree_90:0.0f}')

# theta <= 90
ax2.plot([0, v1[0]], [0, v1[1]], 'b', label='v1')
ax2.plot([0, v3[0]], [0, v3[1]], 'r', label='v3')
ax2.set_xticks(np.arange(-axtick, axtick))
ax2.set_yticks(np.arange(-axtick, axtick))
ax2.axis((-axlim, axlim, -axlim, axlim))
ax2.set_title(f'theta={angle_degree_less_than_90:0.0f}')

# theta >= 90
ax3.plot([0, v1[0]], [0, v1[1]], 'b', label='v1')
ax3.plot([0, v4[0]], [0, v4[1]], 'r', label='v4')
ax3.set_xticks(np.arange(-axtick, axtick))
ax3.set_yticks(np.arange(-axtick, axtick))
ax3.axis((-axlim, axlim, -axlim, axlim))
ax3.set_title(f'theta={angle_degree_greater_than_90:0.0f}')


# theta = 0
ax4.plot([0, v1[0]], [0, v1[1]], 'b', label='v1')
ax4.plot([0, v5[0]], [0, v5[1]], 'r', label='v5')
ax4.set_xticks(np.arange(-axtick, axtick))
ax4.set_yticks(np.arange(-axtick, axtick))
ax4.axis((-axlim, axlim, -axlim, axlim))
ax4.set_title(f'theta={angle_degree_0:0.0f}')


plt.show()
```
